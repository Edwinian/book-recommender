```python
import pandas as pd

books = pd.read_csv('books_cleaned.csv')
```

```python
books["categories"].value_counts().reset_index()
```

```python
category_mapping = {
    "Fiction": "Fiction",
    "Juvenile Fiction": "Children's Fiction",
    "Biography & Autobiography": "Nonfiction",
    "History": "Nonfiction",
    "Literary Criticism": "Nonfiction",
    'Philosophy': 'Nonfiction',
    'Religion': 'Nonfiction',
    'Comics & Graphic Novels': 'Fiction',
    "Drama": "Fiction",
    'Juvenile Nonfiction': "Children's Nonfiction",
    "Science": "Nonfiction",
    "Poetry": "Fiction",
}
```

```python
books["simple_categories"] = books["categories"].map(category_mapping)
```

```python
books
```

```python
# ~ = Not
books[~(books["simple_categories"].isna())]
```

```python
from transformers import pipeline

pipe = pipeline("zero-shot-classification", model="facebook/bart-large-mnli", device="cpu")
# simple_categories = list(set(category_mapping.values()))
simple_categories = ['Fiction', 'Nonfiction']
```

```python
sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[0]
```

```python
result = pipe(sequence, simple_categories)
```

```python
import numpy as np

# np.argmax = find the index of the item with the highest number
max_index = np.argmax(result["scores"])
max_label = result["labels"][max_index]
```

```python
def generate_predictions(sequence, simple_categories):
    result = pipe(sequence, simple_categories)
    max_index = np.argmax(result["scores"])
    max_label = result["labels"][max_index]
    return max_label
```

```python
books
```

```python
from tqdm import tqdm

actual_cats = []
predicted_cats = []

for category in simple_categories:
    for i in tqdm(range(0, 10)):
        sequence = books.loc[books["simple_categories"] == category, "description"].reset_index(drop=True)[i]
        predicted_category = generate_predictions(sequence, simple_categories)
        predicted_cats.append(predicted_category)
        actual_cats.append(category)
```

```python
predicted_df = pd.DataFrame({"actual_cats": actual_cats, "predicted_cats": predicted_cats})
predicted_df
```

```python
predicted_df["correct_prediction"] = np.where(predicted_df["predicted_cats"] == predicted_df["actual_cats"], 1, 0)
predicted_df["correct_prediction"].sum() / len(predicted_df)
```

```python
# Execute predictions on books with missing categories
isbns = []
predicted_cats = []
missing_cats = books.loc[books["simple_categories"].isna(),["isbn13","description"]].reset_index(drop=True)
missing_cats
```

```python
for i in tqdm(range(0, len(missing_cats))):
    sequence = missing_cats["description"][i]
    predicted_cats.append(generate_predictions(sequence, simple_categories))
    isbns.append(missing_cats["isbn13"][i])
```

```python
missing_predicted_df = pd.DataFrame({"isbn13": isbns, "predicted_categories": predicted_cats})
```

```python
# Merge into original DF and replacing missing categories with predicted categories
# Left join on isbn13 col, keeping all rows of existing df
books = pd.merge(books, missing_predicted_df, on="isbn13", how="left")
books["simple_categories"] = np.where(books["simple_categories"].isna(), books["predicted_categories"], books["simple_categories"])
books = books.drop(columns=["predicted_categories"])
```

```python
books.to_csv("books_with_categories.csv", index=False)
```
