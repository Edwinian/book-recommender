```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_chroma import Chroma
from langchain_openai.embeddings import OpenAIEmbeddings
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

books = pd.read_csv('books_cleaned.csv')
```

```python
books['tagged_description']
```

```python
# Langchain text loader does not work directly with panda dataframe, so save tagged descriptions in a text file
# Save tagged_description to a text file, one entry per line
with open('tagged_description.txt', 'w', encoding='utf-8') as f:
    for desc in books['tagged_description']:
        if pd.notna(desc):  # Skip NaN values
            f.write(f"{desc}\n")
```

```python
raw_documents = TextLoader('tagged_description.txt', encoding='utf-8').load()
```

```python
# Set chunk_size to 0 to prioritize splitting by separator over chunk size
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator='\n')
```

```python
documents = text_splitter.split_documents(raw_documents)
```
